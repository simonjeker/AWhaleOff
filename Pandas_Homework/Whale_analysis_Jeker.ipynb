```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline

whales_df = Path('../Resources/whale_returns.csv')
whales_df = pd.read_csv(whales_df, index_col="Date", infer_datetime_format=True, parse_dates=True)
whales_df = whales_df.sort_index()
whales_df.head()
```

```python
#Cou nulls
whales_df.isnull() 
```

```python
# Count mulls
whales_df.isnull().sum()
```

```python
# Dropna function to drop nulls
whales_df.dropna(inplace=True)
whales_df
```

```python
whales_df.isnull().sum()

```

```python
algo_df = Path('../Resources/algo_returns.csv')
algo_df = pd.read_csv(algo_df, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_df = algo_df.sort_index()
algo_df.head()

```

```python
#Identify nulls
algo_df.isnull()
```

```python
# Count mulls
algo_df.isnull().sum()
```

```python
# Dropna function to drop nulls
algo_df.dropna(inplace=True)
algo_df
```

```python
algo_df.isnull().sum()

```

```python
# Reading S&P 500 Closing Prices
sp500_df = Path('../Resources/sp500_history.csv')
sp500_df = pd.read_csv(sp500_df, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500_df = sp500_df.sort_index()
sp500_df.head()
```

```python
#Data type
sp500_df.dtypes
```

```python
sp500_df["Close"] = sp500_df["Close"].str.replace("$", "")
sp500_df["Close"]
```

```python
#Data type
sp500_df.dtypes
```

```python
# Convert `order_total` from `object` to `float`
sp500_df["Close"] = sp500_df["Close"].str.replace('$', '')
sp500_df["Close"] = sp500_df["Close"].astype("float")
sp500_df.dtypes
```

```python
sp500_df.dtypes
```

sp500_df.head()

```python
sp500_df.head()
```

```python
# Calculate the daily return using the 'shift()' function
daily_returns = sp500_df.pct_change()
daily_returns.head()
```

```python
#Drop nulls
daily_returns.dropna(inplace=True)
daily_returns
```

```python
daily_returns = daily_returns.rename(columns={
    "Close": "SP500"
})

daily_returns.head()
```

```python
# Concatenate DataFrames by columns and perform an inner join
combined_returns = pd.concat([whales_df, algo_df, daily_returns], axis="columns", join="inner")
combined_returns
```

```python
combined_returns.plot(figsize=(20,10), title="Daily Returns")
```

```python
cumulative_returns = (1 + combined_returns).cumprod()
cumulative_returns.plot(figsize=(20,10), title="Cumulative Returns")
```

```python
 
#Algo 1 is the clear winner, followed by Berkshire Hathaway INC, the only whale fund beating the S&P500. all other funds were clearly underperformed the S&P500.
#Paulson & CO INC was the only fund continously underperformimg the SP500. #boo

# Box plot to visually show risk

combined_returns.plot.box(figsize=(20, 10), title="Portfolio Risk")
```

```python
#Analysis:
#Tiger Global has the widest spread compared to others. While it's symmetric with a ~short IQR, it has some far spread outliers --Definitely riskier.
#Paulson&Co has the smallest spread, it's symmetrical as well, short IQR and short range of whiskers, too and no crazy outliers. Least risky.
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio. 
# Which portfolios are riskier than the S&P 500?
combined_returns.std()
```

```python
#Riskiest portfolio
sp500_bar = combined_returns["SP500"].std()
combined_returns.std() > sp500_bar
```

```python
#Anual std
combined_returns.std() * np.sqrt(252)
```

```python
#Rolling 21 window
combined_returns.rolling(window=21).std().plot(figsize=(20, 10), title="21 Day Rolling Standard Deviation")
```

```python
#Analysis: I'd say most funds' risk increases with the S&P500 with the exception of Tiger Global which has a few more dramatic spikes in volatility.

#Correlation

correlation = combined_returns.corr()
correlation
```

```python
#Analysis of correlation table: (Not sure I understand it correctly) It seems that Algo 2 at .858... has a very strong correlation to the SP500,
# as well as Soros Fund at .837....

# Beta
covariance = combined_returns["Algo 1"].rolling(window=60).cov(combined_returns["SP500"])
variance = combined_returns["SP500"].rolling(21).var()
(covariance / variance).plot(figsize=(20, 10), title="Algo 1 Beta")
```

```python
#Analysis of Algo 1 Beta: Not sure I'm reading it well, but I don't think it is.
#Expo Weighted Avrg. formula: df['XYZ'].ewm(span= halflife = x) add.std().plot(figsize X) and title
combined_returns.ewm(halflife=21).std().plot(figsize=(20, 10), title="EWA")
```

```python
# Calculate annualized Sharpe Ratios
sharpe_ratios = (combined_returns.mean() * 252) / (combined_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```
